```python
#check the data is normally disributed or not
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure()
sns.histplot(df['price'],kde = True)
plt.title("Histogram + Kde for selling price")
plt.xlabel("selling price")
plt.ylabel("Frequency")
plt.show()
```

```python
# what we need to do if it is not normally distributed?
# -> we are transforming the data by applying log function
import numpy as np
df['log_price'] = np.log(df['price'])
sns.histplot(df['log_price'],kde = True)
plt.show()
```

```python
# Q) Does the fuel type affects the selling price ?
# h0 : No difference in price
# h1 : There is difference in price

#---> which test we will apply t-test or z-test ?
from scipy.stats import ttest_ind
gas = df[df['fueltype']=='gas']['price']
diesel = df[df['fueltype']=='diesel']['price']
stat,p = ttest_ind(gas,diesel)
print("p-value:",p)
#0.05
if p<0.05:
  print("Reject H0")
else:
  print("Accept H1")
```
